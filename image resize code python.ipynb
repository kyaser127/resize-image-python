```python
image = Image.open('compressed.jpg')# image name 
image # printing the image
```

```python
# The file format of the source file.
#print(image.format) # checking the format of image if its jpg leave it 

# The pixel format used by the image. Typical values are "1", "L", "RGB", or "CMYK."
print(image.mode) # Output: RGB

# Image size, in pixels. The size is given as a 2-tuple (width, height).
print(image.size) #current size of image
```

```python
print(image.palette)
```

```python
new_image = image.resize((300,300)) # enter your desired width and hight 
new_image.save('YASIR_kps.jpg') # name at which new resized image got stored

print(image.size) # Output: (hight,width) previous image size
print(new_image.size)  #new image size
```
